## Polynomial models: High Posterior Density (HPD) plots with 95% credible interval

```python
def plot_hpd_plots(d1):
    """
    Plots the HPD plots for the fitted polynomial models
    """
    fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(24, 18), sharex=True, sharey=True)
    degree = 1
    for row in range(0, 3):
        for col in range(0, 2):
            M1, trace1 = load_model_trace(
                filename='fitted_models/M%d.pkl' % degree)
            plot_polynomial_regressions(d1,
                                        trace1,
                                        xcol='age',
                                        ycol='height',
                                        degree=degree,
                                        credible_interval=.95,
                                        ax=ax[row, col])
            degree += 1
```

```python
plot_hpd_plots(df_howel1_train)
```

## Model Evaluation using WAIC information criteria

Here we will compare different models based on their WAIC (Watanabe–Akaike information criterion). These computations are done on the training data that was used to fot the model. We will rank the model based on their WAIC score and the model with lowest WAIC score is our best model.

```python
def compare_waic_for_models():

    model_trace_dict = {}
    for degree in range(1, 7):
        model, trace = load_model_trace(
            'fitted_models/M%d.pkl' %
            (degree))
        model.name = 'M%d' % (degree)
        model_trace_dict[model.name] = trace

    df_comp_WAIC = pm.compare(model_trace_dict)
    pd.options.display.float_format = '{:,.2f}'.format

    print('\n\n')
    print('Table sorted by the WAIC scores \n')
    print(df_comp_WAIC)

    print('------------------------------------------------------------------------\n')
    print('\n\n')
    print('------------------------------------------------------------------------\n')
    print('Displaying the WAIC information on a plot \n')
    pm.compareplot(df_comp_WAIC)
```

```python
compare_waic_for_models()
```

**Conclusion:** The polynomial model with degree 4 seems to be the best model based on the WAIC score.
