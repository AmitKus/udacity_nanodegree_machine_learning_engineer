## Polynomial models: High Posterior Density (HPD) plots with 95% credible interval

```python
def plot_hpd_plots(d1):
    """
    Plots the HPD plots for the fitted polynomial models
    """
    fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(24, 18), sharex=True, sharey=True)
    degree = 1
    for row in range(0, 3):
        for col in range(0, 2):
            M1, trace1 = load_model_trace(
                filename='fitted_models/M%d.pkl' % degree)
            plot_polynomial_regressions(d1,
                                        trace1,
                                        xcol='age',
                                        ycol='height',
                                        degree=degree,
                                        credible_interval=.95,
                                        ax=ax[row, col])
            degree += 1
```

## Model Evaluation using WAIC information criteria

Here we will compare different models based on their WAIC (WatanabeAkaike information criterion). These computations are done on the training data that was used to fot the model. We will rank the model based on their WAIC score and the model with lowest WAIC score is our best model.

**Widely Applicable Information Criterion (WAIC):** It does not require a multivariate Gaussian posterior, and it is often more accurate than DIC. The distinguishing feature of WAIC is that it is pointwise. It access flexibility of a model with respect to fitting each observation, and then sumps up across all observations.

$WAIC = -2 (\sum_{i=1}^N \log Pr(y_i) - \sum_{i=1}^N V(y_i))$

Where $V(y_i)$ is the variance in the log-likelihood for observation $i$ in the training sample.

http://www.stat.columbia.edu/~gelman/research/published/waic_understand3.pdf

logpdf will be imported from stats package. </br>

$p(x) = \frac{1}{\sqrt{ 2 \pi \sigma^2 }} e^{ - \frac{ (x - \mu)^2 } {2 \sigma^2} }$

$log(p(x)) = -0.5*log{2 \pi \sigma^2} - \frac{ (x - \mu)^2 } {2 \sigma^2}$

From WAIC definition (compute for each $y_i$ or weigth data):

$\log Pr(y_i) = -0.5*log{2 \pi \sigma^2} - \frac{ (y_i - \mu_i)^2 } {2 \sigma^2}$

```python
def compare_waic_for_models():

    model_trace_dict = {}
    for degree in range(1, 7):
        model, trace = load_model_trace(
            'fitted_models/M%d.pkl' %
            (degree))
        model.name = 'M%d' % (degree)
        model_trace_dict[model.name] = trace

    df_comp_WAIC = pm.compare(model_trace_dict)
    pd.options.display.float_format = '{:,.2f}'.format

    print('\n\n')
    print('Table sorted by the WAIC scores \n')
    print(df_comp_WAIC)

    print('------------------------------------------------------------------------\n')
    print('\n\n')
    print('------------------------------------------------------------------------\n')
    print('Displaying the WAIC information on a plot \n')
    pm.compareplot(df_comp_WAIC)
```

```python
compare_waic_for_models()
```

**Conclusion:** The polynomial model with degree 4 seems to be the best model based on the WAIC score.

## Model Evaluation using test-sample deviance

WAIC is an estimate of the out-of-sample deviance. Here we will utilize the test data to compute the test sample deviance and compare the models. We will also answer the question if WAIC does a good job in estimating the test deviance.

```python
from scipy.stats import norm
from scipy.special import logsumexp
from sklearn.preprocessing import PolynomialFeatures

def compute_deviance(d1,trace):
    """
    Computes the sample deviance on data d1.
    
    Inputs:
        d1: Dataframe of data with colums 'age' and 'height'
        trace: trace object for fitted model
    """
    #yi
    yi = d1['height'].values

    # mu
    degree = trace['b'].shape[1]
    polynomial_features= PolynomialFeatures(degree=degree, include_bias=False)
    xi = polynomial_features.fit_transform(d1['age'].values.reshape(-1,1))
    a = trace['a'][:,np.newaxis]
    b = trace['b']
    mu = a + np.dot(b,xi.T)
    
    #sigma
    sigma = trace['sigma'] 
    
    # Make a nsamples X nobs log-liklihoods matrix
    loglik = np.zeros(mu.shape)
    nsamples = mu.shape[0]
    nobs = mu.shape[1]
    for i in range(nsamples):
        loglik[i,:] = norm.logpdf(yi.ravel(), loc=mu[i,:].ravel(), scale=sigma[i])
    
    # compute lppd
    lppd = np.zeros((nobs,1))
    for i in range(nobs):
        lppd[i] = logsumexp(loglik[:,i]) - np.log(nsamples)
    
    # computing pwaic
    pwaic = np.zeros((nobs,1))
    for i in range(nobs):
        pwaic[i] = np.var(loglik[:,i])
    
    # finally compute WAIC
    waic = -2*(np.sum(lppd) - np.sum(pwaic))
    return waic

def load_model_trace_dict(ndegree=7):
    """
    loads the saveed model and dict objects.
    """
    model_trace_dict = {}
    for degree in range(1, ndegree):
        model, trace = load_model_trace(
            'fitted_models/M%d.pkl' %
            (degree))
        model.name = 'M%d' % (degree)
        model_trace_dict[model.name] = trace
    return model_trace_dict
        
def compare_models_deviance(df_train,df_test,ndegree=7):
    """
    Computes the sample deviance using the training and test dataset
    
    Input:
        df_train: training data Dataframe
        df_test: test data Dataframe
        ndegree: polynomials upto degree ndegree-1 are uploaded
    """
    model_deviance_dict = {'name':[],'waic':[],'test-sample-deviance':[]}
    model_trace_dict = load_model_trace_dict(ndegree)
    for key,trace in model_trace_dict.items():
        model_deviance_dict['name'].append(key)
        model_deviance_dict['waic'].append(compute_deviance(df_train,trace))
        model_deviance_dict['test-sample-deviance'].append(compute_deviance(df_test,trace))
    
    df_deviance = pd.DataFrame.from_dict(model_deviance_dict)

    return df_deviance
```

```python
# test-sample Deviance
df_deviance = compare_models_deviance(df_howel1_train,df_howel1_test)
print(df_deviance)
```

```python
df_deviance_melt = pd.melt(df_deviance, id_vars=['name'], value_vars=['waic', 'test-sample-deviance'])
sns.catplot(x="name", y="value", hue="variable", data=df_deviance_melt);
```
